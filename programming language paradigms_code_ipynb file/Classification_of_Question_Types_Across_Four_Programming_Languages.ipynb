**Table 6. Classification of Question Types Across Four Programming Languages**

```python
import pandas as pd
import matplotlib.pyplot as plt

# 실제 데이터 입력
data = {
    'Language': ['Python', 'Java', 'C', 'Prolog'],
    'Other Questions (%)': [85, 77, 85, 81],
    'How To Questions (%)': [19, 18, 12, 15],
    'Problem Questions (%)': [2, 2, 2, 2]
}

# DataFrame으로 변환
df = pd.DataFrame(data)

# 히스토그램 생성
df.plot(x='Language', kind='bar', stacked=True)
plt.title('Classification of Question Types Across Four Programming Languages')
plt.xlabel('Programming Language')
plt.ylabel('Percentage (%)')
plt.legend(title='Question Type')
plt.show()

# 바 차트 생성
df.plot(x='Language', kind='barh', stacked=True)
plt.title('Classification of Question Types Across Four Programming Languages')
plt.xlabel('Percentage (%)')
plt.ylabel('Programming Language')
plt.legend(title='Question Type')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 실제 데이터 입력
data = {
    'Language': ['Python', 'Java', 'C', 'Prolog'],
    'Other Questions (%)': [85, 77, 85, 81],
    'How To Questions (%)': [19, 18, 12, 15],
    'Problem Questions (%)': [2, 2, 2, 2]
}

# DataFrame으로 변환
df = pd.DataFrame(data)

# 히스토그램 생성
ax = df.plot(x='Language', kind='bar', stacked=True, figsize=(10, 7))
plt.title('Classification of Question Types Across Four Programming Languages')
plt.xlabel('Programming Language')
plt.ylabel('Percentage (%)')
plt.legend(title='Question Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# 막대 위에 수치 표시
for container in ax.containers:
    ax.bar_label(container, label_type='center')

plt.tight_layout()  # 레이아웃 조정
plt.show()

# 바 차트 생성
ax2 = df.plot(x='Language', kind='barh', stacked=True, figsize=(10, 7))
plt.title('Classification of Question Types Across Four Programming Languages')
plt.xlabel('Percentage (%)')
plt.ylabel('Programming Language')
plt.legend(title='Question Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# 막대 위에 수치 표시
for container in ax2.containers:
    ax2.bar_label(container, label_type='center')

plt.tight_layout()  # 레이아웃 조정
plt.show()
```
