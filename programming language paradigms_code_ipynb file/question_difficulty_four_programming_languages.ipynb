```python
import re

# Read the test.txt file
file_path = '/content/drive/MyDrive/four programming/four programming_Q_list/C_data_Q_list.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Define a function to calculate difficulty based on question complexity
def calculate_difficulty(question):
    # Length of the question
    length_score = len(question)

    # Number of words in the question
    words = re.findall(r'\w+', question)
    word_count = len(words)

    # Average word length
    avg_word_length = sum(len(word) for word in words) / word_count if word_count > 0 else 0

    # Complexity score as a weighted sum of the above factors
    complexity_score = length_score * 0.5 + word_count * 0.3 + avg_word_length * 0.2
    return complexity_score

# Extract the difficulties from each line and print them
difficulties = []
for line in lines:
    question = line.strip()
    if question:  # Ensure the line is not empty
        difficulty = calculate_difficulty(question)
        difficulties.append(difficulty)
        print(f"질문: {question}, 난이도: {difficulty:.2f}")

# Calculate the average difficulty
if difficulties:
    average_difficulty = sum(difficulties) / len(difficulties)
else:
    average_difficulty = 0

print(f'평균 난이도: {average_difficulty:.2f}')
```

```python
import re
import pandas as pd

# Read the test.txt file
file_path = '/content/drive/MyDrive/four programming/four programming_Q_list/java_data_Q_list.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Define a function to calculate difficulty based on question complexity
def calculate_difficulty(question):
    # Length of the question
    length_score = len(question)

    # Number of words in the question
    words = re.findall(r'\w+', question)
    word_count = len(words)

    # Average word length
    avg_word_length = sum(len(word) for word in words) / word_count if word_count > 0 else 0

    # Complexity score as a weighted sum of the above factors
    complexity_score = length_score * 0.5 + word_count * 0.3 + avg_word_length * 0.2
    return complexity_score

# Extract the difficulties from each line and store results
difficulties = []
difficulty_ranges = {
    "10-20": 0,
    "20-30": 0,
    "30-40": 0,
    "40-50": 0,
    "50-60": 0,
    "60-70": 0,
    "70-80": 0,
    "80-90": 0,
    "90-100": 0
}

results = []  # To store question and its difficulty

for line in lines:
    question = line.strip()
    if question:  # Ensure the line is not empty
        difficulty = calculate_difficulty(question)
        difficulties.append(difficulty)
        results.append({"Question": question, "Difficulty": difficulty})

        # Increment the appropriate difficulty range
        if 10 <= difficulty < 20:
            difficulty_ranges["10-20"] += 1
        elif 20 <= difficulty < 30:
            difficulty_ranges["20-30"] += 1
        elif 30 <= difficulty < 40:
            difficulty_ranges["30-40"] += 1
        elif 40 <= difficulty < 50:
            difficulty_ranges["40-50"] += 1
        elif 50 <= difficulty < 60:
            difficulty_ranges["50-60"] += 1
        elif 60 <= difficulty < 70:
            difficulty_ranges["60-70"] += 1
        elif 70 <= difficulty < 80:
            difficulty_ranges["70-80"] += 1
        elif 80 <= difficulty < 90:
            difficulty_ranges["80-90"] += 1
        elif 90 <= difficulty <= 100:
            difficulty_ranges["90-100"] += 1

# Calculate the average difficulty
if difficulties:
    average_difficulty = sum(difficulties) / len(difficulties)
else:
    average_difficulty = 0

# Prepare data for Excel
difficulty_counts = [{"Range": k, "Count": v} for k, v in difficulty_ranges.items()]
average_difficulty_data = [{"Average Difficulty": average_difficulty}]

# Create a Pandas DataFrame
df_questions = pd.DataFrame(results)
df_counts = pd.DataFrame(difficulty_counts)
df_average = pd.DataFrame(average_difficulty_data)

# Create an Excel writer
with pd.ExcelWriter('difficulty_results.xlsx') as writer:
    df_questions.to_excel(writer, sheet_name='Questions', index=False)
    df_counts.to_excel(writer, sheet_name='Difficulty Ranges', index=False)
    df_average.to_excel(writer, sheet_name='Average Difficulty', index=False)

print('결과가 difficulty_results.xlsx 파일로 저장되었습니다.')
```
