 Similarity between data for four programming language questions

```python
import networkx as nx
import matplotlib.pyplot as plt

# 유사성 데이터 입력
similarity_data = {
    ('Python', 'Java'): 0.97,
    ('Python', 'C'): 0.96,
    ('Python', 'Prolog'): 0.91,
    ('Java', 'C'): 0.97,
    ('Java', 'Prolog'): 0.92,
    ('C', 'Prolog'): 0.92
}

# 네트워크 그래프 생성
G = nx.Graph()

# 노드와 엣지 추가
for (lang1, lang2), weight in similarity_data.items():
    G.add_edge(lang1, lang2, weight=weight)

# 노드 위치 설정 (원형 배치)
pos = nx.circular_layout(G)

# 네트워크 그래프 그리기
plt.figure(figsize=(8, 8))
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color='skyblue')
nx.draw_networkx_edges(G, pos, width=2)

# 엣지에 유사성 값 (가중치) 표시
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# 노드에 라벨 표시
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

plt.title('Similarity between Programming Languages')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 유사성 데이터 입력
similarity_data = {
    ('Python', 'Java'): 0.97,
    ('Python', 'C'): 0.96,
    ('Python', 'Prolog'): 0.91,
    ('Java', 'C'): 0.97,
    ('Java', 'Prolog'): 0.92,
    ('C', 'Prolog'): 0.92
}

# 네트워크 그래프 생성
G = nx.Graph()

# 노드와 엣지 추가
for (lang1, lang2), weight in similarity_data.items():
    G.add_edge(lang1, lang2, weight=weight)

# 노드 위치 설정 (원형 배치)
pos = nx.circular_layout(G)

# 네트워크 그래프 그리기
plt.figure(figsize=(8, 8))
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color='skyblue')
nx.draw_networkx_edges(G, pos, width=2)

# 엣지에 유사성 값 (가중치) 표시 - label_pos 값을 0.3으로 설정하여 레이블 위치 조정
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, label_pos=0.3)

# 노드에 라벨 표시
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

plt.title('Similarity between Programming Languages')
plt.show()
```
